```python
#Tree based model
# 1. 데이터 표준화 X // 계산 X, 질문하는식 학습
# 2. 성능 굳
# 3. 데이터 설명력 가짐
# 4.
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

import pandas as pd
```

```python
df = pd.read_csv('fish2.csv')
X = df[["Weight","Length", "Diagonal", "Height","Width"]].values
Y = df[["Species"]].values

train_x, test_x, train_y, test_y = train_test_split(X, Y) # 75: 25 default
```

```python
model = DecisionTreeClassifier()
model.fit(train_x, train_y)
```

```python
model.score(test_x, test_y)
```

```python
# 분류: 정확도 , 정밀도 , 재현율
# 정밀도 =>
from sklearn.metrics import classification_report
pred = model.predict(test_x)

report = classification_report(test_y, pred)
print(report)
```

```python
import seaborn as sns

iris = sns.load_dataset('iris') # 3종류의 꽃
iris
```

```python
A = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
B = iris[['species']].values

train_a, test_a, train_b, test_b = train_test_split(A, B)
```

```python
model = DecisionTreeClassifier()
model.fit(train_a, train_b)
model.score(test_a, test_b)
```

```python
pred = model.predict(test_a) # Test ma'lumotlari (test_a) yordamida modeldan bashoratlar olamiz.

report = classification_report(test_b, pred) # Haqiqiy qiymatlar (test_b) va bashorat qilingan qiymatlar (pred) asosida klassifikatsiya hisobotini yaratamiz.
print(report) # Klassifikatsiya hisobotini chop etamiz.
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plot_tree(model, feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], filled=True)
plt.show()
```
